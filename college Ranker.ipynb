## Task 2: Visualizing the Target Distribution

```python
sns.distplot (df['revenue'])
plt.show()
```

```python
#visualizing log transformed revenue distribution
sns.displot(np.log1p(df['revenue']))
plt.show()
```

## Task 3: Relationship between Film Revenue and Budget

```python
sns.scatterplot(df['revenue'],df['budget'])
plt.title("Relationship between film revenue and budget")
plt.show()
```

```python
#relation using corr matrix lets find ot they are corelated or not
relation=df[['revenue','budget']].corr()
sns.heatmap(relation,vmin=0,vmax=1,center=0)
plt.show()
```
